```python
from parrot import Parrot
import torch
import warnings
import csv

def random_state(seed):
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

random_state(1234)

parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=True)

dataset_path = "MRP_Dataset_Final_Non_Paraphrased.csv"
paraphrased_dataset = []

with open(dataset_path, "r") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        question, answer = row[0], row[1]

        print("-" * 100)
        print("Original Question:", question)
        print("-" * 100)

        para_phrases = parrot.augment(input_phrase=question, max_return_phrases=10)

        if para_phrases is not None:
            for para_phrase in para_phrases:
                paraphrased_question = para_phrase[0]
                paraphrased_dataset.append((paraphrased_question, answer))
        else:
            # Handle the case when para_phrases is None
            print("No paraphrased phrases found for the question:", question)

        # Store the original question and answer
        paraphrased_dataset.append((question, answer))

# Save the paraphrased dataset to a new CSV file
output_path = "paraphrased_dataset.csv"

with open(output_path, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(paraphrased_dataset)
```
